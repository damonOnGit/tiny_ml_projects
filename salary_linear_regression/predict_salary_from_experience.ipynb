```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./Salary_dataset.csv')
df = df.drop(columns=df.columns[0])
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
from sklearn import linear_model, model_selection
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

reg = linear_model.LinearRegression()
training_df, testing_df = model_selection.train_test_split(df)

reg.fit(training_df[['YearsExperience']], training_df['Salary'])
print(f'Intercept: {reg.intercept_}\nCoefficients: {reg.coef_}')

predictions = reg.predict(testing_df[['YearsExperience']])

# Mean Absolute Error (MAE) – Average absolute difference between actual and predicted values.
# Mean Squared Error (MSE) – Average squared difference (penalizes larger errors more).
# R² Score (Coefficient of Determination) – Measures how well the model explains variance in data (1 = perfect fit, 0 = no relationship).
mae = mean_absolute_error(testing_df['Salary'], predictions)
mse = mean_squared_error(testing_df['Salary'], predictions)
r2 = r2_score(testing_df['Salary'], predictions)


print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'R² Score: {r2}')
```
